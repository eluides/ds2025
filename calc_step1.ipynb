```python
from decimal import Decimal, getcontext
import ipywidgets as widgets
from IPython.display import display, clear_output
getcontext().prec = 28 

MIN_VALUE = Decimal('-1000000000000.000000')
MAX_VALUE = Decimal('1000000000000.000000')

def parse_number(input_str):
    normalized = input_str.replace(',', '.')
    try:
        return Decimal(normalized)
    except:
        raise ValueError("Некорректное число")

label_info = widgets.Label(value="Крепская Анна Валерьевна, Курс: 4, Группа: 4, Год: 2025")  
input_a = widgets.Text(value='0.0', description='Число A:')
input_b = widgets.Text(value='0.0', description='Число B:')

operation = widgets.Dropdown(
    options=['Сложение', 'Вычитание'],
    value='Сложение',
    description='Операция:'
)

button_calculate = widgets.Button(description='Вычислить')
output_result = widgets.Output()

def calculate(b):
    with output_result:
        clear_output()
        try:
            a = parse_number(input_a.value)
            b = parse_number(input_b.value)
            
            if operation.value == 'Сложение':
                result = a + b
            else:
                result = a - b
            
            if result < MIN_VALUE or result > MAX_VALUE:
                print("Переполнение: результат вне диапазона")
            else:
                print(f"Результат: {result}")
        except ValueError as e:
            print(f"Ошибка: {e}")
button_calculate.on_click(calculate)

display(label_info, input_a, input_b, operation, button_calculate, output_result)
```
